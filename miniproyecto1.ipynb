Tasks:

2. Considere que usted analizará hasta un máximo de 16 huracanes este año. GrafiquePMF(probabilitymass function) de estos eventos

```python
factorial = lambda n: np.math.factorial(n)
    
def poisson(x, lamb):
    return (lamb**x)*(np.e**(-lamb)) / factorial(x)
```

```python
poisson_pmf = [(i, poisson(i, 7) ) for i in range(0, 17)]
poisson_pmf
```

```python
plt.figure(figsize=(20, 8))
plt.plot([i[0] for i in poisson_pmf], [i[1] for i in poisson_pmf], 'ro')
plt.plot([i[0] for i in poisson_pmf], [i[1] for i in poisson_pmf], color = 'red', linewidth = 2, linestyle = 'solid')
```

```python
print('Suma de probabilidades:', sum([n[1] for n in poisson_pmf]))
```

3. Considere que usted analizará hasta un máximo de 16 huracanes este año. Grafique CDF (cumulative distribution function) de estos eventos

```python
poisson_cdf = []

for n in range(len(poisson_pmf)):
    if n == 0:
        poisson_cdf.append((poisson_pmf[n][0], poisson_pmf[n][1]))
    else:
        poisson_cdf.append((poisson_pmf[n][0], poisson_pmf[n][1] + poisson_cdf[n-1][1]))

poisson_cdf
```

```python
plt.figure(figsize=(20, 8))
plt.plot([i[0] for i in poisson_cdf], [i[1] for i in poisson_cdf], 'ro')
plt.plot([i[0] for i in poisson_cdf], [i[1] for i in poisson_cdf], color = 'red', linewidth = 2, linestyle = 'solid')
```

## Ejercicio 3

Un aspecto importante de los procesos de Poisson es que los intervalos de tiempo entre eventos consecutivos puede ser modelado usando una distribución exponencial, a pesar de que esta últimaseacontínua. Considerando esto para generar tiempos intermedios en un proceso de Poisson, usamos la técnica de invertir la CDF, en la cual literalmente se construye la inversa de la CDF, y se le da como input diferentes valores de una distribución uniforme. Esto da como resultado los correspondientestiempos intermedios con sus respectivas probabilidades.

```python
def cdf_inv(t, lamb):
    return -(np.log(1-t)/lamb)

def cdf_inv_exp(t, lamb):
    return 1 - np.e**(-(lamb*t))
```

1. Genere una tabla que muestre los tiempos intermedios en horas para los 10 primeros pacientes

```python
import random
random.seed(42)

lamb = 5 # por hora

results = [(n, cdf_inv(random.random(), lamb)) for n in range(1, 11)]
results_cdf = []

for n in range(len(results)):
    if n == 0:
        results_cdf.append((results[n][0], results[n][1]))
    else:
        results_cdf.append((results[n][0], results[n][1] + results_cdf[n-1][1]))

results_cdf
```

```python
import tabulate

table = tabulate.tabulate(results_cdf, headers=['n', 'cdf_inv'], tablefmt='fancy_grid')
print(table)
```

## Ejercicio 5

Considere un experimento en el cual se cuentan cuántos carros pasan por una calle determinada dentro de un
rango de tiempo dado. Sabemos que el tiempo de espera para el n-ésimo evento puede ser modelando a través de
una variable gamma. Considere el caso en el que usted está esperando que pase el 3er carro en la calle dada.


1. Para diferentes lambdas en [2, 1, 0.5], grafique las distribuciones gamma para cada uno de los casos

```python
K = 3
t1 = 0.5
t2 = 1
t3 = 2
```

```python
x = np.arange(0,20,0.1)
```

```python
y1 = (x**(K-1) * np.exp((-1) * (x/t1)))/((t1**K)*math.factorial(K-1))
y2 = (x**(K-1) * np.exp((-1) * (x/t2)))/((t2**K)*math.factorial(K-1))
y3 = (x**(K-1) * np.exp((-1) * (x/t3)))/((t3**K)*math.factorial(K-1))
```

```python
plt.title("Distribuciones Gamma (k = 3)")
plt.plot(x,y1,label="λ = 2",color='red')
plt.plot(x,y2,label="λ = 1",color='green')
plt.plot(x,y3,label="λ = 0.5",color='blue')
plt.legend(bbox_to_anchor=(1,1), loc='upper right', borderaxespad=1)
plt.ylim([0,0.6])
plt.show()
```
